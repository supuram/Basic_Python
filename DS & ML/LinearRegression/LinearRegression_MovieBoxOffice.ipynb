```python
import pandas
from pandas import DataFrame
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = pandas.read_csv('cost_revenue_dirty.csv')
```

```python
data.describe()
```

```python
X = DataFrame(data,columns=['production_budget_usd'])
y = DataFrame(data,columns=['worldwide_gross_usd'])
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X,y,alpha=0.3)
plt.title('Film Cost vs Global Revenue')
plt.xlabel('Production Budget $')
plt.ylabel('WorldWide Gross $')
plt.xlim(0,450000000)
plt.ylim(0,3000000000)
plt.show()
```

```python
regression = LinearRegression()
regression.fit(X,y)
```

```python
regression.coef_  
#Here slope of the line is getting calculated. It says that for every dollar we spend on movie we can hope to make 3.11 dollar back in return
```

```python
regression.intercept_
#Here the intercept of the line is getting calculated. Here the intercept of -70lakh is saying that a movie with a budge of $0 would lose 70lakh dollars. Now thats a bit problematic.
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X,y,alpha=0.3)
plt.plot(X,regression.predict(X),color='red',linewidth=4)  #X is the actual values of Production_Budget and predict(X) is the predicted value from regression using slope and intercept
plt.title('Film Cost vs Global Revenue')
plt.xlabel('Production Budget $')
plt.ylabel('WorldWide Gross $')
plt.xlim(0,450000000)
plt.ylim(0,3000000000)
plt.show()
```

```python
regression.score(X,y)
#This is also called goodness of fit or r^2 or R^2. This number represents the amount of variation in film revenue that is explained by the film's budget. That is from slope we got that revenue will be 3.11 times the movies budget. This will come true 55% of the times as from the data here.
```
