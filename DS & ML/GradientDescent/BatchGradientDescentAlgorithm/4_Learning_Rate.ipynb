## The Learning Rate
## $$g(x) = x^4 - 4x^2 + 5$$

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_2 = np.linspace(-2, 2, 1000)

def g(x):
    return x**4 - 4*x**2 + 5

def dg(x):
    return 4*x**3 - 8*x
```

```python
def gradient_descent(derivative_func, initial_guess, multiplier=0.02, precision=0.001, max_iteration=300):
    new_x = initial_guess
    x_list = [new_x]
    slope_list = [derivative_func(new_x)]

    for n in range(max_iteration):
        previous_x = new_x
        gradient = derivative_func(previous_x)
        new_x = previous_x - multiplier * gradient
        step_size = abs(new_x - previous_x)
        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))
        if step_size < precision:
            break
    return new_x, x_list, slope_list
    
```

```python
plt.figure(figsize = [15,4])
local_min, list_x, deriv_list = gradient_descent(derivative_func=dg, initial_guess=1.9, multiplier=0.25, 
max_iteration=5)

#1st plot
plt.subplot(1,2,1) #1 indicates 1 row, 2 indicates 2 columns
plt.title('Cost Function')
plt.xlabel('x')
plt.ylabel('g(x)')
plt.plot(x_2, g(x_2))
plt.scatter(list_x, g(np.array(list_x)), color="red")

#2nd plot
plt.subplot(1,2,2)
plt.title('Slope of Cost Function')
plt.xlabel('x')
plt.ylabel(r'$\frac{dg(x)}{dx}$') #The r before the string indicates that it is a raw string, which means that backslashes will be treated as literal backslashes rather than escape characters.
plt.plot(x_2, dg(x_2))
plt.scatter(list_x, deriv_list, color="red")
plt.show()
print('Number of steps is ', len(list_x))
```

## Effect of Learning Rate on Cost

```python
n=100
plt.figure(figsize = [20,4])
low_gamma = gradient_descent(derivative_func=dg, initial_guess=3, multiplier=0.0005, 
precision=0.0001, max_iteration=n)
mid_gamma = gradient_descent(derivative_func=dg, initial_guess=3, multiplier=0.001, 
precision=0.0001, max_iteration=n)
high_gamma = gradient_descent(derivative_func=dg, initial_guess=3, multiplier=0.002, 
precision=0.0001, max_iteration=n)
insane_gamma = gradient_descent(derivative_func=dg, initial_guess=1.9, multiplier=0.25, 
precision=0.0001, max_iteration=n)

#1st plot
plt.subplot(1,2,1) #1 indicates 1 row, 2 indicates 2 columns
plt.title('Effect of the Learning Rate')
plt.xlim(0,n)
plt.ylim(0, 50)
plt.xlabel('No of iterations')
plt.ylabel('g(x)')

#X Axis data
iteration_list = list(range(0, n+1))

#Y Axis data
low_values = np.array(low_gamma[1])
mid_values = np.array(mid_gamma[1])
high_values = np.array(high_gamma[1])
insane_values = np.array(insane_gamma[1])

plt.plot(iteration_list, g(low_values), color="blue")
plt.scatter(iteration_list, g(low_values), color="blue", s=10)

plt.plot(iteration_list, g(mid_values), color="red")
plt.scatter(iteration_list, g(mid_values), color="red", s=10)

plt.plot(iteration_list, g(high_values), color="green")
plt.scatter(iteration_list, g(high_values), color="green", s=10)

plt.plot(iteration_list, g(insane_values), color="hotpink")
plt.scatter(iteration_list, g(insane_values), color="hotpink", s=10)

plt.show()
```
