# Data Visualization with 3D
## Minimise $$f(x,y) = \frac{1}{3^{-x^2-y^2}+1}$$
Minimise $$f(x,y) = \frac{1}{r+1}$$ where $r$ is ${3^{-x^2-y^2}}$

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import cm
```

```python
def f(x, y):
    r = 3 ** (-x ** 2 - y ** 2)
    return 1/(r + 1)
```

```python
x = np.linspace(start=-1, stop=2, num=200)
y = np.linspace(start=-1, stop=2, num=200)

print('Shape of x array ', x.shape)
print('Shape of y array ', y.shape)
```

## Transforming x and y to a 2D array to plot on 3D

```python
x = np.linspace(start=-1, stop=2, num=200)
y = np.linspace(start=-1, stop=2, num=200)
x, y = np.meshgrid(x, y)
print('Array after meshgrid ', x.shape)
print('Array after meshgrid ', y.shape)
```

```python
fig = plt.figure(figsize=[26,8])
axis = fig.add_subplot(111, projection='3d')
axis.set_xlabel('X', fontsize=20)
axis.set_ylabel('Y', fontsize=20)
axis.set_zlabel('f(x,y) - Cost', fontsize=20)
axis.plot_surface(x, y, f(x, y), cmap=cm.coolwarm)
plt.show()
```
