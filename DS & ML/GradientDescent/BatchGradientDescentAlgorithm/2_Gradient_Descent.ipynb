```python
import matplotlib.pyplot as plt
import numpy as np
```

# Example2 (Multiple Minima vs Initial Guess and Advanced Functions)
## $$g(x) = x^4 - 4x^2 + 5 $$

```python
x_2 = np.linspace(-2, 2, 1000)

def g(x):
    return x**4 - 4*x**2 + 5

def dg(x):
    return 4*x**3 - 8*x
```

```python
plt.figure(figsize = [15,4])

#1st plot
plt.subplot(1,2,1) #1 indicates 1 row, 2 indicates 2 columns
plt.title('Cost Function')
plt.xlabel('x')
plt.ylabel('g(x)')
plt.plot(x_2, g(x_2))

#2nd plot
plt.subplot(1,2,2)
plt.title('Slope of Cost Function')
plt.xlabel('x')
plt.ylabel(r'$\frac{dg(x)}{dx}$') #The r before the string indicates that it is a raw string, which means that backslashes will be treated as literal backslashes rather than escape characters.
plt.plot(x_2, dg(x_2))
plt.show()
```

## Gradient Descent as a Python Function
The thing about Python is that a function is actually a full blown object. So functions are stored in a piece of the computer's memory all on their own, just like other objects are.
And this means that you can stick a function in a variable and pass functions around our program.
So in our gradient descent function this derivative_func parameter will be our place holder for the actual derivative function and you'll see this when we call this function.

```python
def gradient_descent(derivative_func, initial_guess, multiplier=0.02, precision=0.001):
    new_x = initial_guess
    x_list = [new_x]
    slope_list = [derivative_func(new_x)]

    for n in range(500):
        previous_x = new_x
        gradient = derivative_func(previous_x)
        new_x = previous_x - multiplier * gradient
        step_size = abs(new_x - previous_x)
        
        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))
        if step_size < precision:
            break
    return new_x, x_list, slope_list
    
```

```python
local_min, list_x, deriv_list = gradient_descent(dg, 0.5, 0.02, 0.001)
print('Local min occurs at ', local_min)
print('Number of steps is ', len(list_x))
```

```python
local_min, list_x, deriv_list = gradient_descent(derivative_func=dg, initial_guess=-0.5, multiplier=0.01, 
precision=0.0001)
print('Local min occurs at ', local_min)
print('Number of steps is ', len(list_x))
```

```python
local_min, list_x, deriv_list = gradient_descent(derivative_func=dg, initial_guess=-0.1)
print('Local min occurs at ', local_min)
print('Number of steps is ', len(list_x))
```

```python
plt.figure(figsize = [15,4])
local_min, list_x, deriv_list = gradient_descent(derivative_func=dg, initial_guess=0.1)

#1st plot
plt.subplot(1,2,1) #1 indicates 1 row, 2 indicates 2 columns
plt.title('Cost Function')
plt.xlabel('x')
plt.ylabel('g(x)')
plt.plot(x_2, g(x_2))
plt.scatter(list_x, g(np.array(list_x)), color="red")

#2nd plot
plt.subplot(1,2,2)
plt.title('Slope of Cost Function')
plt.xlabel('x')
plt.ylabel(r'$\frac{dg(x)}{dx}$') #The r before the string indicates that it is a raw string, which means that backslashes will be treated as literal backslashes rather than escape characters.
plt.plot(x_2, dg(x_2))
plt.scatter(list_x, deriv_list, color="red")
plt.show()
```

Lets now feed our initial_guess as 0. We end up sitting right here on the maximum. And that's because the slope at this very, very point is also equal to 0. And we can see this here on the right hand function here, on the slope of the cost function. Remember, a gradient descent algorithm stops running once the slope is equal to 0.
