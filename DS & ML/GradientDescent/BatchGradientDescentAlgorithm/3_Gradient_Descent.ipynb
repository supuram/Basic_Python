```python
import matplotlib.pyplot as plt
import numpy as np
```

## Example3 - Divergence, Overflow
## $$h(x) = x^5 - 2x^4 + 2$$

```python
x_3 = np.linspace(-2.5, 2.5, 1000)

def h(x):
    return x**5 - 2*x**4 + 2

def dh(x):
    return 5*x**4 - 8*x**3
```

```python
plt.figure(figsize = [15,4])

#1st plot
plt.subplot(1,2,1) #1 indicates 1 row, 2 indicates 2 columns
plt.title('Cost Function')
plt.xlabel('x')
plt.ylabel('h(x)')
plt.plot(x_3, h(x_3))

#2nd plot
plt.subplot(1,2,2)
plt.title('Slope of Cost Function')
plt.xlabel('x')
plt.ylabel(r'$\frac{dh(x)}{dx}$') #The r before the string indicates that it is a raw string, which means that backslashes will be treated as literal backslashes rather than escape characters.
plt.plot(x_3, dh(x_3))
plt.show()
```

```python
def gradient_descent(derivative_func, initial_guess, multiplier=0.02, precision=0.001, max_iteration=300):
    new_x = initial_guess
    x_list = [new_x]
    slope_list = [derivative_func(new_x)]

    for n in range(max_iteration):
        previous_x = new_x
        gradient = derivative_func(previous_x)
        new_x = previous_x - multiplier * gradient
        step_size = abs(new_x - previous_x)
        
        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))
        if step_size < precision:
            break
    return new_x, x_list, slope_list
```

```python
local_min, list_x, deriv_list = gradient_descent(dh, 0.5, 0.02, 0.001)
print('Local min occurs at ', local_min)
print('Number of steps is ', len(list_x))
```

```python
plt.figure(figsize = [15,6])
local_min, list_x, deriv_list = gradient_descent(derivative_func=dh, initial_guess=0.2)

#1st plot
plt.subplot(1,2,1) #1 indicates 1 row, 2 indicates 2 columns
plt.title('Cost Function')
plt.xlabel('x')
plt.ylabel('g(x)')
plt.xlim(-1.5, 2.5)
plt.ylim(-1, 4)
plt.plot(x_3, h(x_3))
plt.scatter(list_x, h(np.array(list_x)), color="red")

#2nd plot
plt.subplot(1,2,2)
plt.title('Slope of Cost Function')
plt.xlabel('x')
plt.ylabel(r'$\frac{dg(x)}{dx}$') #The r before the string indicates that it is a raw string, which means that backslashes will be treated as literal backslashes rather than escape characters.
plt.xlim(-0.5, 1.7)
plt.ylim((-4, 5))
plt.plot(x_3, dh(x_3))
plt.scatter(list_x, deriv_list, color="red")
plt.show()

print('Local min occurs at ', local_min)
print('Cost at this minimum is ', h(local_min))
print('Number of steps', len(list_x))
```
