```python
import matplotlib.pyplot as plt
import numpy as np
```

# Example1
$f(x) = x^2 + x + 1$

```python
def f(x):
    return x**2 + x + 1
```

```python
x_1 = np.linspace(-3 , 3, 1000)
```

```python
plt.plot(x_1, f(x_1))
plt.show()
```

# Finding the value of x that minimises f(x)

Slopes and Derivatives

```python
#Derivative of f(x)
def df(x):
    return 2*x + 1
```

```python
plt.figure(figsize = [10,4])

#1st plot
plt.subplot(1,2,1) #1 indicates 1 row, 2 indicates 2 columns
plt.title('Cost Function')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.plot(x_1, f(x_1))

#2nd plot
plt.subplot(1,2,2)
plt.title('Slope of Cost Function')
plt.xlabel('x')
plt.ylabel(r'$\frac{df(x)}{dx}$') #The r before the string indicates that it is a raw string, which means that backslashes will be treated as literal backslashes rather than escape characters.
plt.plot(x_1, df(x_1))
plt.show()
```

## Gradient Descent

```python
new_x = -3
previous_x = 0
step_multiplier = 0.1
precision = 0.00001

x_list = [new_x]
slope_list = [df(new_x)]

for n in range(500):
    previous_x = new_x
    gradient = df(previous_x)
    new_x = previous_x - step_multiplier * gradient
    step_size = abs(new_x - previous_x)
    
    x_list.append(new_x)
    slope_list.append(df(new_x))
    if step_size < precision:
        break
print('Local Minimum = ', new_x)
print('Value of df(x) = ', df(new_x))
print('f(x) value or cost at this point = ', f(new_x))
    
```

```python
plt.figure(figsize = [20,4])

#1st plot
plt.subplot(1,3,1) 
plt.title('Cost Function')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.plot(x_1, f(x_1), color="red")

values = np.array(x_list)
plt.scatter(x_list, f(values), s=50, alpha=0.9)

#2nd plot
plt.subplot(1,3,2)
plt.title('Slope of Cost Function')
plt.xlabel('x')
plt.ylabel(r'$\frac{df(x)}{dx}$') #The r before the string indicates that it is a raw string, which means that backslashes will be treated as literal backslashes rather than escape characters.
plt.grid()
plt.plot(x_1, df(x_1), color="red")
plt.scatter(x_list, slope_list, s=50, alpha=0.9)

#3rd plot
plt.subplot(1,3,3)
plt.title('Gradient Descent(close up)')
plt.xlabel('x')
plt.ylabel(r'$\frac{df(x)}{dx}$') #The r before the string indicates that it is a raw string, which means that backslashes will be treated as literal backslashes rather than escape characters.
plt.grid()
plt.xlim(-0.55, -0.2)
plt.ylim(-0.3, 0.8)
plt.plot(x_1, df(x_1), color="red")
plt.scatter(x_list, slope_list, s=100, alpha=0.9)
plt.show()
```
